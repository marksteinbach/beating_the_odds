### The Naive Model 

The naive model assumes that the teams score according to a normal distribution with a mean equal to arithmetic mean of Points Per Game (PPG) and Opponent Points Allowed Per Game (OppPAPG) and a variance equal to the sample variance of the PPG.

```python
### This is the naive model
mavs = np.random.normal((A_B + ybar_A)/2, np.sqrt(s2_A), 1000)
spurs = np.random.normal((A_A + ybar_B)/2, np.sqrt(s2_B), 1000)

print "The naive chance of the Mavs winning is " + str(sum(mavs > spurs)/1000.0) + "%"
```

Let's create a better model, that can scale according to how much we want to take into account the prior (Defensive efficiency) versus the data (Scoring per game)

Let $$s_{A,B} | \theta_{A,B}, \sigma^2  \sim N(\theta_{A,B}, \sigma^2)$$ be the number of points team A will score on team B.

Then, $$\theta_{A,B} | \sigma^2 \sim N(A_B, \sigma^2) $$ where $A_B$ is the number of points allowed by team B per game. 

And finally, $$\sigma^2 \sim Inverse-Gamma(v_0/2, v_0\sigma^2_0/2) $$

The posterior $\theta_{A,B} | y, \sigma^2$ is therefore distributed $$\theta_{A,B} | y, \sigma^2 \sim N(\mu_{A,B}, \sigma^2/k_A)$$ and $$\sigma^2 | y \sim Inverse-Gamma(v_n/2, v_n \sigma^2_n / 2),$$ where
$$\mu_{A,B} = (k_0 + A_B + n_A \bar{y}_A)/(k_0 + n_A) $$
$$k_A = k_0 + n_A$$
$$v_n = v_0 + n_A + n_B$$
$$\sigma^2_{n,A} = 1/v_n[v_0\sigma^2_0 + (n_A - 1)s^2_A + k_0{n_A \over k_A}(\bar{y}_A - A_B)^2]$$

We will be able to tune the $v_0$ and $k_0$ parameters using training data to find the ideal values for betting purposes. The intuition behind these parameters is that they are the weights for prior and data. See below for a sensitivity analysis on $v_0$ and$k_0$

```python
### A slightly more complicated model -- multi-level Bayesian model
# Given more time and computational resources one should tune v_0 and k_0 by season.
k_0 = 20.00
v_0 = 20.00
sigma2_0 = 100.
nsim = 1000


# Refer to math above to follow this code
k_A = k_0 + n_A
k_B = k_0 + n_B
mu_AB = (k_0*A_B + n_A*ybar_A)/(k_A)
mu_BA = (k_0*A_A + n_B*ybar_B)/(k_B)
v_n = v_0 + n_A + n_B
sigma2_n_A = (1/v_n)*(v_0*sigma2_0 + (n_A - 1)*s2_A + k_0*n_A/k_A*(ybar_A - A_B)**2)
sigma2_n_B = (1/v_n)*(v_0*sigma2_0 + (n_B - 1)*s2_B + k_0*n_B/k_B*(ybar_B - A_A)**2)
sigma2_AB = 1/np.random.gamma(v_0/2, 1/(sigma2_n_A*v_0/2), nsim)
sigma2_BA = 1/np.random.gamma(v_0/2, 1/(sigma2_n_B*v_0/2), nsim)
theta_AB = np.random.normal((mu_AB), np.sqrt(sigma2_AB), nsim)
theta_BA = np.random.normal((mu_BA), np.sqrt(sigma2_BA), nsim)
s_AB = np.random.normal(theta_AB, np.sqrt(sigma2_AB), nsim)
s_BA = np.random.normal(theta_BA, np.sqrt(sigma2_BA), nsim)

### Fun little histogram that explores insight into over/under betting.
plt.hist(s_AB + s_BA)
plt.title("Distribution of combined score in Mavs-Spurs game")
plt.ylabel("Frequency")
plt.xlabel("Combined Score")
plt.show()
```

```python
#Sensitivity analysis 
nsim = 10000
v_0 = np.floor(range(1,60,1))
k_0 = v_0
prob_s = [1]*len(k_0)
prob_theta = [1]*len(k_0)

for i in range(len(v_0)):
    
    #refer to math above to follow this code
    k_A = k_0[i] + n_A
    k_B = k_0[i] + n_B
    mu_AB = (k_0[i]*A_B + n_A*ybar_A)/(k_A)
    mu_BA = (k_0[i]*A_A + n_B*ybar_B)/(k_B)
    v_n = v_0[i] + n_A + n_B
    sigma2_n_A = (1/v_n)*(v_0[i]*sigma2_0 + (n_A - 1)*s2_A + k_0[i]*n_A/k_A*(ybar_A - A_B)**2)
    sigma2_n_B = (1/v_n)*(v_0[i]*sigma2_0 + (n_B - 1)*s2_B + k_0[i]*n_B/k_B*(ybar_B - A_A)**2)
    sigma2_AB = 1/np.random.gamma(v_0[i]/2, 1/(sigma2_n_A*v_0[i]/2), nsim)
    sigma2_BA = 1/np.random.gamma(v_0[i]/2, 1/(sigma2_n_B*v_0[i]/2), nsim)
    theta_AB = np.random.normal((mu_AB), np.sqrt(sigma2_AB), nsim)
    theta_BA = np.random.normal((mu_BA), np.sqrt(sigma2_BA), nsim)
    s_AB = np.random.normal(theta_AB, np.sqrt(sigma2_AB), nsim)
    s_BA = np.random.normal(theta_BA, np.sqrt(sigma2_BA), nsim)
    prob_s[i] = sum(s_AB > s_BA)/np.floor(nsim)
    prob_theta[i] = sum(theta_AB > theta_BA)/np.floor(nsim)
```

### Data Wrangling

The original box_scores dataframe is not ideal for our model. Here we wrangle the box_scores dataset into a new dataset, toFeed, that will be fed into our whatToBet function. Because a single game is represented in two separate lines in the original dataframe, this code uses that data to build a new dataframe in which the information for both teams is contained in a single row. It is designed so that the team favored to win by Vegas is always designated as 'Team A'. It follows, then, that all betting lines for the spread of the game are negative.

```python
grouped = box_scores.groupby('Game ID')

teamA = []
teamB = []
date = []
teamAscore = []
teamBscore = []
sprOpen = []
sprOpenOdds = []
totalOpen = []
totalOpenOdds = []

# loosely typed datasets cause problems somteimes. this function is helpful for quick checks

def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False

# gathers data from grouped box scores table and organizes it into lists
for name, subdf in grouped:
    if len(subdf) > 1:
        subdf = subdf.reset_index()
        
        # we construct our dataset so that team A is the team preferred by Vegas
        if isfloat(subdf['Spread Open'][0]):
            date.append(subdf['Date'][0])
            if float(subdf['Spread Open'][0]) < 0:
                a = 0
                b = 1

            else:
                a = 1
                b = 0

            teamA.append(subdf['Teams'][a])
            teamB.append(subdf['Teams'][b])

            teamAscore.append(int(subdf['Scores'][a]))
            teamBscore.append(int(subdf['Scores'][b]))

            sprOpen.append(float(subdf['Spread Open'][a]))
            sprOpenOdds.append(float(subdf['Spr Open Odds'][a]))
            totalOpen.append(float(subdf['Total Open'][a]))
            totalOpenOdds.append(float(subdf['Ttl Open Odds'][a]))
        
# creates organized dataframe of box score information to be fed to betting model
toFeed = pd.DataFrame({'Date': date, 'Team A': teamA, 'Team B':teamB,
                       'Team A Score':teamAscore, 'Team B Score':teamBscore,
                       'Spread Open':sprOpen, 'Spr Open Odds':sprOpenOdds,
                       'Total Open':totalOpen, 'Ttl Open Odds':totalOpenOdds})

toFeed.head()
```

### Helper functions

We define a function that accepts a date and a team. It returns a list of the points that team has scored in all games leading up to the given date in that given date's season. The output of this function will be a parameter for the whatToBet function.

```python
# function that creates a list of a team's game-by-game point totals through a certain date in 
# the season
def getScores(date, team):
    
    season = (date + datetime.timedelta(80)).year
    thisYear = box_scores[(box_scores['Teams'] == team) & (box_scores['Season']== season)]
    thisYear = thisYear.reset_index()
    prevGameIndices = np.where(np.array([(date - day).days for day in thisYear['Date']])>0)[0]
    prevGames = thisYear.loc[prevGameIndices,:]
    
    scores = thisYear['Scores'][prevGameIndices]
    
    return scores
```

Next, we define a similar function with identical parameters that returns a list of the points conceded by the passed team in each game leading up to the given date in that given date's season. The average of this function's output will be a parameter for the whatToBet function.

```python
# function that creates a list of a team's game-by-game points allowed through a 
# certain date in the season
def getConcededPoints (date, team):
    season = (date + datetime.timedelta(80)).year
    thisYear = box_scores[(box_scores['Teams'] == team) & (box_scores['Season']== season)]

    thisYear = thisYear.reset_index()
    prevGameIndices = np.where(np.array([(date - day).days for day in thisYear['Date']])>0)[0]

    prevGames = thisYear.loc[prevGameIndices,:]

    otherTeams = box_scores[(box_scores['Teams'] != team) & (box_scores['Season'] == season)]
    opponentResults = otherTeams.set_index('Game ID').loc[prevGames['Game ID'],:]

    return opponentResults['Scores']
```

## Refining the Bayesian model to a function

```python
def whatToBet(team1scoreList, team2scoreList, team1avgConceded, team2avgConceded, OverUnder, line):
    
    OverUnder = float(OverUnder)
    line = float(line)
    
    A = team1scoreList
    B = team2scoreList

    # Create values holding the length of A and B
    n_A = np.floor(len(A))
    n_B = np.floor(len(B))

    # Create values holding the sample mean and variance of A and B 
    ybar_A = np.mean(A)
    ybar_B = np.mean(B)
    s2_A = np.var(A)
    s2_B = np.var(B)

    # each team's points allowed per game 
    A_A = team1avgConceded
    A_B = team2avgConceded
    
    # refer to the math above to follow this code
    k_0 = 20.00
    v_0 = 20.00
    sigma2_0 = 100.
    nsim = 2000

    k_A = k_0 + n_A
    k_B = k_0 + n_B
    mu_AB = (k_0*A_B + n_A*ybar_A)/(k_A)
    mu_BA = (k_0*A_A + n_B*ybar_B)/(k_B)
    v_n = v_0 + n_A + n_B
    sigma2_n_A = (1/v_n)*(v_0*sigma2_0 + (n_A - 1)*s2_A + k_0*n_A/k_A*(ybar_A - A_B)**2)
    sigma2_n_B = (1/v_n)*(v_0*sigma2_0 + (n_B - 1)*s2_B + k_0*n_B/k_B*(ybar_B - A_A)**2)
    sigma2_AB = 1/np.random.gamma(v_0/2, 1/(sigma2_n_A*v_0/2), nsim)
    sigma2_BA = 1/np.random.gamma(v_0/2, 1/(sigma2_n_B*v_0/2), nsim)
    theta_AB = np.random.normal((mu_AB), np.sqrt(sigma2_AB), nsim)
    theta_BA = np.random.normal((mu_BA), np.sqrt(sigma2_BA), nsim)
    s_AB = np.random.normal(theta_AB, np.sqrt(sigma2_AB), nsim)
    s_BA = np.random.normal(theta_BA, np.sqrt(sigma2_BA), nsim)
    
    probOver = sum(s_AB + s_BA > OverUnder)/float(nsim)
    
    # determines the Over/under bet based on a minimum confidence margin of 55% (inspired by odds set at -110 by Vegas)
    if (probOver > 0.55):
        OUbet = 'O'
    elif (probOver < 0.45):
        OUbet = 'U'
    else:
        OUbet = 'None'
    
    probOverLine = sum(s_AB - s_BA > line)/float(nsim)
    
    # determines the spread bet based on a minimum confidence margin of 55% (inspired by odds set at -110 by Vegas)
    if (probOverLine > 0.55):
        lineBet = 'A'
    elif (probOverLine < 0.5):
        lineBet = 'B'
    else:
        lineBet = 'None'
    
    # return the results in the form of a dictionary
    return {'LineBet':lineBet, 'OUBet':OUbet}
```

```python
# function to call whatToBet based on input data from the table toFeed
# it's also designed to not call whatToBet unnecessarily for months we plan to drop

def bet(teamA, teamB, date, OverUnder, line):
    
    # the burn in period where the model does not bet
    # this functionality improves our model but also our computational efficiency
    if date.month == 10 or date.month == 11 or date.month == 12 or date.month == 1 or date.month == 2:
        return {'LineBet':'None', 'OUBet':'None'}
    
    # if the burn in period has passed, get the appropriate wagering recommendation
    else:
        teamAscores = list(getScores(date, teamA))
        teamBscores = list(getScores(date, teamB))

        team1conceded = np.mean(getConcededPoints(date, teamA))
        team2conceded = np.mean(getConcededPoints(date, teamB))

        return whatToBet(teamAscores, teamBscores, team1conceded, team2conceded, OverUnder, line)
```
